```python
# Now, let's perform further data cleaning by setting the data types of all 
# columns for consistency in our dataframe.

# We set 'Home' to boolean values to signify whether a game record corresponds 
# to a home game or an away game.

# We also set 'Team1Won' to booleans to represent the result of each game.

from datetime import datetime

formatted_df = trimmed_df
formatted_df['Date'] = formatted_df['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))

numeric_columns = ['ID',
				   'Season',
				   'Team1Pts',    
					 'Team2Pts',    
					 'Team1PtDiff',
					 'Team2PtDiff', 
					 'Team1TM',    
					 'Team2TM',     
					 'Team1Rating', 
					 'Team2Rating', 
					 'Team1Sks',    
					 'Team2Sks',    
					 'Team1SkYds',  
					 'Team2SkYds',  
					 'Team1RushAtt',
					 'Team2RushAtt',
					 'Team1RushYds',
					 'Team2RushYds',
					 'Team1RYM',    
					 'Team2RYM',    
					 'Team1PYM',    
					 'Team2PYM',    
					 'Team1YM',     
					 'Team2YM']

str_columns = ['Team1',
			   'Team2']

for column in numeric_columns:
	formatted_df[column] = pd.to_numeric(formatted_df[column], errors='coerce')

for column in str_columns:
	formatted_df[column] = formatted_df[column].astype(str)

formatted_df['Home'] = formatted_df['Home'].isnull()

formatted_df['Team1Won'] = formatted_df['Team1Won'].apply(lambda x: True if str(x).startswith('W') else False)

for column in formatted_df:
	print(type(formatted_df[column][0]))
```

```python
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# List of columns to exclude for training purposes
exclude = ['ID', 'Team1Won', 'Season', 'Date', 'Team1', 'Team2']  # Columns to exclude

# Feature matrix X and target vector y
X = formatted_df.drop(columns=exclude)  # Drop the excluded columns
y = formatted_df['Team1Won']  # Target variable

# List of pre-game and post-game columns based on the provided dataset
pre_game_columns = ['Team1', 'Team2', 'Home']
post_game_columns = ['Team1Pts', 'Team2Pts', 'Team1RushYds', 'Team2RushYds', 'Team1SkYds', 'Team2SkYds',
                     'Team1Sks', 'Team2Sks', 'Team1RushAtt', 'Team2RushAtt', 'Team1RYM', 'Team2RYM', 
                     'Team1PYM', 'Team2PYM', 'Team1YM', 'Team2YM', 'Team1Rating', 'Team2Rating']  # Including ratings here
outcome_column = 'Team1Won'  # Assuming this is your target column

# Function to create filler data for post-game features using historical averages (last 5 games from training set)
def generate_filler_data(team, full_data, train_data, num_games=5):
    # Filter the train_data for the games involving the given team
    team_data = train_data[(train_data['Team1'] == team) | (train_data['Team2'] == team)].sort_values('Date', ascending=False)
    
    # Ensure only past games from the training data are considered (avoid data from future test games)
    if len(team_data) >= num_games:
        filler_data = team_data.head(num_games).mean(axis=0)
    else:
        filler_data = team_data.mean(axis=0)  # Use whatever data is available if less than num_games
    return filler_data

# Function to prepare training and testing datasets for each fold of TimeSeriesSplit
def prepare_data_for_fold(fold, full_data, num_games=5):
    # Split data into training and testing based on the fold
    train_indices, test_indices = fold
    train_data = full_data.iloc[train_indices]
    test_data = full_data.iloc[test_indices]
    
    # Extract pre-game features
    X_train = train_data[pre_game_columns]
    X_test = test_data[pre_game_columns]  # Only use pre-game features during testing
    
    # Generate filler data for post-game features
    X_test_filled = X_test.copy()
    for index, row in X_test.iterrows():
        team1 = row['Team1']  # Get Team1
        team2 = row['Team2']  # Get Team2
        
        # Generate filler data for both teams (Team1 and Team2) from the training set only
        filler_data_team1 = generate_filler_data(team1, full_data, train_data, num_games)
        filler_data_team2 = generate_filler_data(team2, full_data, train_data, num_games)
        
        # Replace missing post-game features for Team1 with filler data
        for col in post_game_columns:
            if 'Team1' in col:  # Fill Team1's post-game columns
                X_test_filled.at[index, col] = filler_data_team1[col]
            elif 'Team2' in col:  # Fill Team2's post-game columns
                X_test_filled.at[index, col] = filler_data_team2[col]
    
    # Extract the outcome data for training and testing
    y_train = train_data[outcome_column]
    y_test = test_data[outcome_column]
    
    return X_train, X_test_filled, y_train, y_test

# Define the hyperparameter grid for the decision tree
param_grid = {
    'max_depth': [5, 10, 15, 20],
    'min_samples_split': [2, 5, 10]
}

# Initialize the DecisionTreeClassifier
dt = DecisionTreeClassifier()

# Set up TimeSeriesSplit for cross-validation
tscv = TimeSeriesSplit(n_splits=5)

# Store results from each fold
all_accuracy_scores = []

# Nested cross-validation
for train_indices, test_indices in tscv.split(X):  # Use X, which is the feature matrix for TimeSeriesSplit
    # Prepare training and testing data for this fold
    X_train, X_test_filled, y_train, y_test = prepare_data_for_fold(
        (train_indices, test_indices), full_data, num_games=5
    )
    
    # GridSearchCV (Inner Loop for Hyperparameter Tuning)
    grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=tscv)
    grid_search.fit(X_train, y_train)
    
    # Get the best model from grid search
    best_model = grid_search.best_estimator_

    # Test the model on the testing data
    y_pred = best_model.predict(X_test_filled)
    
    # Calculate accuracy and store the results
    accuracy = accuracy_score(y_test, y_pred)
    all_accuracy_scores.append(accuracy)

    # Optionally, print out the best hyperparameters for each fold
    print(f"Best parameters for this fold: {grid_search.best_params_}")

# Calculate and print average accuracy across all folds
average_accuracy = sum(all_accuracy_scores) / len(all_accuracy_scores)
print(f"Average accuracy across all folds: {average_accuracy:.4f}")
```
